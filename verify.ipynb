```python
import matplotlib.pyplot as plt

n_4_tower = []
n_5_tower = []
n = []

def calculateCoefN4():
    res = []
    for n in range(1, 11):
        res.append(n*(n + 1))
    return res

def calculateCoefN5():
    res = []
    for n in range(1, 11):
        res.append(n*n)
    return res

n_4_direct = calculateCoefN4()
n_5_direct = calculateCoefN5()

with open('tower_res.txt', 'r') as f:
    for line_no, line in enumerate(f):
        if line_no % 2 == 0:
            n.append(int(line[2:]))
        else:
            coef = line.split(' ')
            n_4_tower.append(0 - int(coef[-3]))
            n_5_tower.append(0 - int(coef[-4]))
                       
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111)

plt.xlabel("n")
plt.ylabel("Number of combinations")

ax.plot(n, n_4_tower, 'g-', label="n + 4 (Tower)")
ax.plot(n, n_5_tower, 'r-', label="n + 5 (Tower)")
ax.plot(n, n_4_direct, 'bo', label="n + 4 (directed)")
ax.plot(n, n_5_direct, 'co', label="n + 4 (directed)")

plt.legend()
plt.title('Number of combinations of lattice paths by segments usage in 2 x n network')

plt.savefig('verify.jpg')
plt.show()
                       
```
